```python
def garaje(var):
    var = str(var).lower()
    var = var.replace(" ","").replace("cub.","cubierto").replace("desc.","descubierto")
    if 'a' in var:
        return 'comunal'
    elif 'no' in var:
        return 'no'
    elif var=='0':
        return 'no'
    elif var=='2':
        return 'doble'
    elif var=='3':
        return 'triple'
    elif var=='4':
        return 'cuatruple'
    elif var=='incluído':
        return 'sencillo'
    elif var=='si':
        return 'sencillo'
    elif var=='1':
        return 'sencillo'
    else:
        return var
```

```python
df['garaje'] = df['garaje'].apply(garaje)
```
